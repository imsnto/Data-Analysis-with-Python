```python
auto = pd.read_csv('/content/Automobile (1).csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['normalized_losses'])
plt.show()
```

```python
sns.distplot(auto['normalized_losses'], kde=False, rug=True)
plt.show()
```

```python
sns.jointplot(x='engine_size', y='horsepower', data=auto) # Pass data as keyword argument and x and y as strings.
plt.show()
```

```python
sns.jointplot(x=auto['engine_size'], y=auto['horsepower'], kind='hex') # Pass x and y as keyword arguments.
plt.show()
```

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])
```

```python
sns.stripplot(x=auto['fuel_type'], y=auto['horsepower'], jitter = True)
```

```python
sns.swarmplot(x=auto['fuel_type'], y=auto['horsepower'])
```

```python
sns.boxplot(x=auto['number_of_doors'], y=auto['horsepower'], hue=auto['fuel_type']) # Pass x and y as keyword arguments.
```

```python
sns.barplot(x=auto['body_style'], y=auto['horsepower'], hue= auto['engine_location']) # Pass x and y as keyword arguments.
```

```python
sns.countplot(x=auto['body_style'], hue=auto['engine_location'])
sns.countplot(x='body_style', hue='engine_location', data=auto)
```

```python
sns.pointplot(x=auto['fuel_system'], y=auto['horsepower'], hue=auto['number_of_doors']) # Pass x and y as keyword arguments.
```

```python
sns.lmplot(x='horsepower', y ='peak_rpm', data= auto, hue='fuel_type')
```
